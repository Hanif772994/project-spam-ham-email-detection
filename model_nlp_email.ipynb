```python
import pandas as pd
import numpy as np
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB
```

**Data collec**

```python
df=pd.read_csv("/content/emails.csv")
```

```python
df.head()
```

```python
df["spam"].value_counts()
```

**Duplicate remove**

```python
df.duplicated().sum()
```

```python
df.drop_duplicates(inplace=True)
```

```python
df["spam"].value_counts()
```

```python
df.isnull().sum()
```

**Separate x and y**

```python
x= df.text.values
y= df.spam.values
```

**split dataset**

```python
xtrain, xtest, ytrain, ytest=train_test_split(x,y, test_size=0.2, random_state=42)
```

**Text Preprocessing**

```python
cv=CountVectorizer()
x_train=cv.fit_transform(xtrain)
```

```python
x_train.toarray()
```

**ML Algorithms**

```python
mb=MultinomialNB()
```

```python
model=mb.fit(x_train, ytrain)
```

```python
model.score(x_train, ytrain)
```

```python
x_test=cv.transform(xtest)
```

```python
x_test.toarray()
```

```python
model.score(x_test, ytest)
```

**model testing**

```python
emails=["If you're not seeing your Spam or Trash, go to the bottom of your label list on the left, click More,  then scroll again until you find them ", "I have successfully submitted my assignment for the Machine Learning course."]
```

```python
test=cv.transform(emails)
```

```python
model.predict(test)
```
